```python
import os
from email.message import EmailMessage
import ssl                                # zabezpieczenie maila  
import smtplib
import pandas as pd
```

```python
def df_to_txt(df):
    ''' Przerabia dane z ramki danych na wiadomość z ofertami. '''

    message = 'Oto oferty, które mogą cię zainteresować: \n'
    
    for i in range(df.shape[0]):
        
        mess = '''
                Lokalizacja: {lokalizacja}
                Cena: {cena}
                Metraż: {powierzchnia}
                Liczba pokoi: {pokoje}
                Piętro: {pietro}
                Rodzaj zabudowy: {r_zabudowy}
                Link:
                '''.format(length = 'multi-line', lokalizacja = df.iloc[i]['lokalizacja'], cena = df.iloc[i]['cena'], powierzchnia = df.iloc[i]['powierzchnia'], \
                    pokoje = df.iloc[i]['pokoje'], pietro = df.iloc[i]['piętro'], r_zabudowy = df.iloc[i]['rodzaj zabudowy'])

        # print(mess)
        message += mess
       
    return message
```

```python
def sent_mail(receiver, df):
    ''' Wysyła maila z ofertami spełniającymi kryteria użytkownika '''

    email_sender = 'piotr.granieczny26@gmail.com'
    email_password = os.environ['Mail_Password'] # ukrywa hasło
    email_receiver = receiver
    
    subject = 'Python Projekt'
    
    body = df_to_txt(df)
    
    em = EmailMessage()
    em['From'] = email_sender
    em['To'] = email_receiver
    em['Subject'] = subject
    em.set_content(body)
    
    context = ssl.create_default_context()  # zabezpieczenie maila 
    
    with smtplib.SMTP_SSL('smtp.gmail.com', 465, context = context) as smtp:
        smtp.login(email_sender, email_password)
        smtp.sendmail(email_sender, email_receiver, em.as_string())  # em.as_string() - zapewnia poprawny format maila
```
