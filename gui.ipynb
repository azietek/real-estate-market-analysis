```python
import tkinter as tk
from tkinter import ttk

class WyszukiwarkaNieruchomosciGUI: 
    def __init__(self, root):
        self.root = root
        self.root.title("Wyszukiwarka Nieruchomości")

        # Tworzenie etykiet i pól do wprowadzania danych
        self.label_lokalizacja = ttk.Label(root, text="Lokalizacja:")
        self.label_lokalizacja.grid(row=0, column=0, padx=10, pady=10, sticky=tk.W)

        self.lokalizacja_var = tk.StringVar()
        self.combo_lokalizacja = ttk.Combobox(root, textvariable=self.lokalizacja_var, values=["Psie Pole", "Krzyki", "Fabryczna", "Śródmieście", "Stare Miasto"])
        self.combo_lokalizacja.grid(row=0, column=1, padx=10, pady=10, sticky=tk.W)

        self.label_cena = ttk.Label(root, text="Cena za (max):")
        self.label_cena.grid(row=1, column=0, padx=10, pady=10, sticky=tk.W)

        self.entry_cena_max = ttk.Entry(root)
        self.entry_cena_max.grid(row=1, column=1, padx=10, pady=10, sticky=tk.W)

        self.label_powierzchnia = ttk.Label(root, text="Powierzchnia (od - do):")
        self.label_powierzchnia.grid(row=2, column=0, padx=10, pady=10, sticky=tk.W)

        self.entry_powierzchnia_od = ttk.Entry(root)
        self.entry_powierzchnia_od.grid(row=2, column=1, padx=10, pady=10, sticky=tk.W)

        self.entry_powierzchnia_do = ttk.Entry(root)
        self.entry_powierzchnia_do.grid(row=2, column=2, padx=10, pady=10, sticky=tk.W)

        self.label_pokoje = ttk.Label(root, text="Liczba pokoi:")
        self.label_pokoje.grid(row=3, column=0, padx=10, pady=10, sticky=tk.W)

        self.pokoje_var = tk.StringVar()
        self.combo_pokoje = ttk.Combobox(root, textvariable=self.pokoje_var, values=["1", "2", "3", "4"])
        self.combo_pokoje.grid(row=3, column=1, padx=10, pady=10, sticky=tk.W)

        self.label_rynek = ttk.Label(root, text="Rynek:")
        self.label_rynek.grid(row=4, column=0, padx=10, pady=10, sticky=tk.W)

        self.rynek_var = tk.StringVar()
        self.combo_rynek = ttk.Combobox(root, textvariable=self.rynek_var, values=["Pierwotny", "Wtórny"])
        self.combo_rynek.grid(row=4, column=1, padx=10, pady=10, sticky=tk.W)

        self.label_stan_wykonczenia = ttk.Label(root, text="Stan wykonczenia:")
        self.label_stan_wykonczenia.grid(row=5, column=0, padx=10, pady=10, sticky=tk.W)

        self.stan_wykonczenia_var = tk.StringVar()
        self.combo_stan_wykonczenia = ttk.Combobox(root, textvariable=self.stan_wykonczenia_var, values=["Do remontu", 'Do wykończenia', "Do zamieszkania"])
        self.combo_stan_wykonczenia.grid(row=5, column=1, padx=10, pady=10, sticky=tk.W)

        # Opcje zaznaczania
        self.label_sortowanie = ttk.Label(root, text="Ważne dla Ciebie cechy (max 3):")
        self.label_sortowanie.grid(row=7, column=0, padx=10, pady=10, sticky=tk.W)

        # Use a BooleanVar for each Checkbutton
        self.sortowanie_var_cena = tk.BooleanVar()
        self.check_cena = ttk.Checkbutton(root, text="Cena", variable=self.sortowanie_var_cena)
        self.check_cena.grid(row=7, column=1, padx=10, pady=10, sticky=tk.W)

        self.sortowanie_var_powierzchnia = tk.BooleanVar()
        self.check_powierzchnia = ttk.Checkbutton(root, text="Powierzchnia", variable=self.sortowanie_var_powierzchnia)
        self.check_powierzchnia.grid(row=7, column=2, padx=10, pady=10, sticky=tk.W)

        self.sortowanie_var_pokoje = tk.BooleanVar()
        self.check_pokoje = ttk.Checkbutton(root, text="Liczba pokoi", variable=self.sortowanie_var_pokoje)
        self.check_pokoje.grid(row=8, column=0, padx=10, pady=10, sticky=tk.W)

        self.sortowanie_var_rynek = tk.BooleanVar()
        self.check_rynek = ttk.Checkbutton(root, text="Rynek", variable=self.sortowanie_var_rynek)
        self.check_rynek.grid(row=8, column=1, padx=10, pady=10, sticky=tk.W)

        self.sortowanie_var_lokalizacja = tk.BooleanVar()
        self.check_lokalizacja = ttk.Checkbutton(root, text="Lokalizacja", variable=self.sortowanie_var_lokalizacja)
        self.check_lokalizacja.grid(row=8, column=2, padx=10, pady=10, sticky=tk.W)

        self.sortowanie_var_stan_wykonczenia = tk.BooleanVar()
        self.check_stan_wykonczenia = ttk.Checkbutton(root, text="Stan wykonczenia", variable=self.sortowanie_var_stan_wykonczenia)
        self.check_stan_wykonczenia.grid(row=9, column=0, padx=10, pady=10, sticky=tk.W)

        self.sortowanie_var_balkon = tk.BooleanVar()
        self.check_balkon = ttk.Checkbutton(root, text="Balkon", variable=self.sortowanie_var_balkon)
        self.check_balkon.grid(row=9, column=1, padx=10, pady=10, sticky=tk.W)

        # Przycisk do uruchamiania wyszukiwania
        self.button_szukaj = ttk.Button(root, text="Szukaj", command=self.wyszukaj_nieruchomosci)
        self.button_szukaj.grid(row=10, column=0, columnspan=3, pady=20)

    def wyszukaj_nieruchomosci(self):
        # Tutaj możesz dodać kod do przetwarzania wprowadzonych danych, np. wysłanie zapytania do bazy danych
        # i wyświetlenie wyników na ekranie lub w innym formacie.

        # Poniżej znajduje się przykład wyświetlenia wprowadzonych danych w konsoli.
        print("Lokalizacja:", self.lokalizacja_var.get())
        print("Cena (max):", self.entry_cena_max.get())
        print("Powierzchnia (od - do):", self.entry_powierzchnia_od.get(), "-", self.entry_powierzchnia_do.get())
        print("Liczba pokoi:", self.pokoje_var.get())
        print("Rynek:", self.rynek_var.get())
        print("Stan wykonczenia:", self.stan_wykonczenia_var.get())

        # Dodanie opcji zaznaczania
        print("Ważne dla Ciebie cechy (max 3):")
        if self.sortowanie_var_cena.get():
            print("- Cena")
        if self.sortowanie_var_powierzchnia.get():
            print("- Powierzchnia")
        if self.sortowanie_var_pokoje.get():
            print("- Liczba pokoi")
        if self.sortowanie_var_rynek.get():
            print("- Rynek")
        if self.sortowanie_var_lokalizacja.get():
            print("- Lokalizacja")
        if self.sortowanie_var_stan_wykonczenia.get():
            print("- Stan wykonczenia")
        if self.sortowanie_var_balkon.get():
            print("- Balkon")

        self.root.destroy()

if __name__ == "__main__":
    root = tk.Tk()
    app = WyszukiwarkaNieruchomosciGUI(root)
    root.mainloop()
```
