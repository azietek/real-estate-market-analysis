```python
def sonar(url,oferta):
    
    driver = webdriver.Chrome()
    driver.get(url)
    driver.maximize_window()
    button = driver.find_element(By.XPATH, '//*[@id="__next"]/aside[1]/div/div[2]/button')
    button.click()
    time.sleep(1+random.random())
    if oferta['powierzchnia'] is not None:
        metraz = driver.find_element(By.XPATH, '//*[@id="input--size"]')
        metraz.send_keys(str(oferta['powierzchnia']))
    time.sleep(1-random.random())
    if oferta['pokoje'] is not None:
        pokoje = driver.find_element(By.XPATH, '//*[@id="input--roomsCount"]')
        pokoje.send_keys(str(oferta['pokoje']))
    time.sleep(random.random())    
    if oferta['piętro'] is not None:
        pietro = driver.find_element(By.XPATH, '//*[@id="input--floor"]')
        pietro.send_keys(str(oferta['piętro']))
    time.sleep(0.65+random.random())    
    miasto = driver.find_element(By.XPATH, '//*[@id="cityId-input"]')
    miasto.send_keys("Wrocław")
    time.sleep(1.23+random.random())
    if oferta['liczba pięter'] is not None:
        liczba_pieter = driver.find_element(By.XPATH, '//*[@id="input--buildingFloorsCount"]')
        liczba_pieter.send_keys(str(oferta['liczba pięter']))
    time.sleep(0.786+random.random())
    if oferta['rok budowy'] is not None:
        rok = driver.find_element(By.XPATH, '//*[@id="input--yearBuilt"]')
        rok.send_keys(str(oferta['rok budowy']))
    time.sleep(random.random())    
    if oferta['balkon/ogrod/taras']==True:
        balkon = driver.find_element(By.XPATH, '//*[@id="new_estimate"]/fieldset/div[3]/div/div[4]/label/span')
        balkon.click()
    time.sleep(1+random.random())    
    if oferta['miejsce parkingowe']==True:
        parking = driver.find_element(By.XPATH, '//*[@id="new_estimate"]/fieldset/div[3]/div/div[3]/label/span')
        parking.click()
    time.sleep(1+3*random.random())   
    if oferta['ulica'] is not None:
        ulica = driver.find_element(By.XPATH, '//*[@id="buildingId-input"]')
        ul = str(oferta['ulica']) + ' 1'
        ulica.send_keys(str(ul))
        time.sleep(1.73)
        ulica.send_keys(Keys.ENTER)

    time.sleep(1-random.random())    
    if oferta['stan wykonczenia']=='Do remontu':
        standard = driver.find_element(By.XPATH, '//*[@id="input--standard"]/option[2]')
        standard.click()
        
    if oferta['stan wykonczenia']=='Do odświeżenia':
        standard = driver.find_element(By.XPATH, '//*[@id="input--standard"]/option[3]')
        standard.click()
        
    if oferta['stan wykonczenia']=='Gotowe do wprowadzenia':
        standard = driver.find_element(By.XPATH, '//*[@id="input--standard"]/option[4]')
        standard.click()
        
    if oferta['stan wykonczenia']=='Wysoki standard/luksusowe':
        standard = driver.find_element(By.XPATH, '//*[@id="input--standard"]/option[5]')
        standard.click()
    
    time.sleep(2)
    ciekawosc = driver.find_element(By.XPATH, '//*[@id="new_estimate"]/fieldset/div[4]/div[1]/label/span')
    ciekawosc.click()
    wyceniam = driver.find_element(By.XPATH, '//*[@id="new_estimate"]/fieldset/button')
    wyceniam.click()
    time.sleep(15)
    try: 
        cena = driver.find_element(By.XPATH, '//*[@id="__next"]/div/div[2]/div[2]/main/div/div/div[1]/div[3]/div[1]') 
        zametr = driver.find_element(By.XPATH, '//*[@id="__next"]/div/div[2]/div[2]/main/div/div/div[1]/div[3]/div[2]') 
    except:
        return None
    
    time.sleep(2.47 + 0.1*random.random())
    driver.quit()
    return [cena.text,zametr.text]
        
```
